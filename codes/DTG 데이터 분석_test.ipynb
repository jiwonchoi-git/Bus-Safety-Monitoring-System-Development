## 특정 데이터만 모아서 따로 보기

```python
#"트립시작일"이 'ts[0]'에 해당하는 데이터만 가져와서 변수에 담기
#copy()를 해주면 df_ts0 변수에 변화가 생겨도 df 원본은 변하지 않기 때문에 안전하다.
df_ts0 = df[df["트립시작일"] == ts[0]].copy()
df_ts0
```

```python
for i in range(len(ts)):
    df_ts = df[df["트립시작일"] == ts[i]].copy()
    print(df_ts)
```
